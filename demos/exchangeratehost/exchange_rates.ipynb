# ExchangeRate.host Demo

This notebook demonstrates how to use the ExchangeRate.host MCP tools with a LangChain agent to get exchange rates and convert currencies.

```python
import os

os.environ["EXCHANGERATEHOST_ACCESS_KEY"] = "your_api_key_here"

from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_openai import ChatOpenAI

from pymcpx.exchangeratehost import MCP_TOOLS
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
agent = create_tool_calling_agent(llm, MCP_TOOLS)
executor = AgentExecutor(
    agent=agent,
    tools=MCP_TOOLS,
    verbose=True,
    handle_parsing_errors=True,
)
```

```python
response = executor.invoke({"input": "Get the latest USD to EUR exchange rate"})
print(response["output"])
```

```python
response = executor.invoke({"input": "Convert 100 USD to GBP using historical rates from 2020-01-01"})
print(response["output"])
```

```python
response = executor.invoke({"input": "Show me EUR/USD rates from 2023-01-01 to 2023-01-07"})
print(response["output"])
```
