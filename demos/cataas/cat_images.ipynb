````python
from dotenv import load_dotenv

load_dotenv()

from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

from demos.shared.llm_setup import get_llm
from pymcpx.cataas import CataasToolkit

llm = get_llm()

tools = CataasToolkit().get_tools()

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a cat image assistant. Use the cataas tools to find cat images "
            "and information. When you get an image URL, present it as a markdown "
            "image link so the user can view it.\n\n"
            "Available tools:\n"
            "- list_tags(search, limit) — search cat tags\n"
            "- list_cats(limit, tags, search) — list cats\n"
            "- get_random_cat(transform) — random cat image URL\n"
            "- get_random_cat_saying(text, transform, text_overlay) — cat saying text",
        ),
        ("human", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
````

````python
result = executor.invoke(
    {"input": "What tags are available for cute cats? Search for 'cute' tags."}
)
print(result["output"])
````

````python
result = executor.invoke({"input": "Show me a random cat image with width 400."})
print(result["output"])
````

````python
result = executor.invoke({"input": "Find me a cute cat that says 'Hello there!'"})
print(result["output"])
````
