```python
from dotenv import load_dotenv
load_dotenv()

from demos.shared.llm_setup import get_llm
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from pymcpx.calculator import (
    AbsTool,
    ModuloTool,
    PowerTool,
    SqrtTool,
)

llm = get_llm()

tools = [PowerTool(), ModuloTool(), SqrtTool(), AbsTool()]

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a helpful math assistant. Use the calculator tools to solve "
        "problems step by step.\n\n"
        "Available tools:\n"
        "- power_numbers(a: float, b: float) \u2014 raise a to power b (a ** b)\n"
        "- modulo_numbers(a: float, b: float) \u2014 remainder of a / b (error if b=0)\n"
        "- sqrt_number(a: float) \u2014 square root of a (error if a < 0)\n"
        "- abs_number(a: float) \u2014 absolute value of a",
    ),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke({"input": "What is 2 raised to the power of 10?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "What is the remainder of 100 divided by 7?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "What is the square root of 144?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "What is the absolute value of -42? What about the absolute value of 3.14?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "Try to calculate the square root of -9. What error do you get?"})
print(result["output"])
```
