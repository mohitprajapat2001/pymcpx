````python
from dotenv import load_dotenv
load_dotenv()

from demos.shared.llm_setup import get_llm
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from pymcpx.calculator import (
    CalculatorToolkit,
)

llm = get_llm()

tools = CalculatorToolkit.get_tools()

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a helpful math assistant. Use the calculator tools to solve "
        "arithmetic problems step by step.\n\n"
        "Available tools:\n"
        "- add_numbers(a: float, b: float) \u2014 add two numbers\n"
        "- subtract_numbers(a: float, b: float) \u2014 subtract b from a\n"
        "- multiply_numbers(a: float, b: float) \u2014 multiply two numbers\n"
        "- divide_numbers(a: float, b: float) \u2014 divide a by b (error if b=0)",
    ),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
````

````python
result = executor.invoke({"input": "What is 15 + 27?"})
print(result["output"])
````

````python
result = executor.invoke({"input": "Subtract 42 from 100."})
print(result["output"])
````

````python
result = executor.invoke({"input": "Multiply 12 by 8."})
print(result["output"])
````

````python
result = executor.invoke({"input": "Divide 144 by 12. Then divide 10 by 0 and tell me what happens."})
print(result["output"])
````

````python
result = executor.invoke({"input": "Calculate the average of 10, 20, and 30."})
print(result["output"])
````
