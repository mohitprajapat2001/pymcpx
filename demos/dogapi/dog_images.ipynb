````python
from dotenv import load_dotenv

load_dotenv()

from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

from demos.shared.llm_setup import get_llm
from pymcpx.dogapi import DogapiToolkit

llm = get_llm()

tools = DogapiToolkit().get_tools()

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a dog breed assistant. Use the dogapi tools to answer questions "
            "about dog breeds and provide dog images. When you get an image URL, "
            "present it as a markdown image link.\n\n"
            "Available tools:\n"
            "- list_breeds(search, limit) — search dog breeds\n"
            "- get_random_dog_image() — random dog image URL\n"
            "- get_random_dog_image_by_breed(breed) — random dog image by breed",
        ),
        ("human", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
````

````python
result = executor.invoke(
    {"input": "What dog breeds are similar to 'hound'? Search for hound-related breeds."}
)
print(result["output"])
````

````python
result = executor.invoke({"input": "Show me a random dog image!"})
print(result["output"])
````

````python
result = executor.invoke({"input": "Find me a random husky image!"})
print(result["output"])
````
