````python
from dotenv import load_dotenv

load_dotenv()

from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

from demos.shared.llm_setup import get_llm
from pymcpx.catfacts import CatfactsToolkit

llm = get_llm()

tools = CatfactsToolkit().get_tools()

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a cat expert assistant. Use the catfacts tools to answer questions "
            "about cat breeds and provide interesting cat facts.\n\n"
            "Available tools:\n"
            "- get_breeds(limit) — list cat breeds\n"
            "- get_random_fact(max_length) — random cat fact\n"
            "- get_facts(limit, max_length) — list multiple facts",
        ),
        ("human", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
````

````python
result = executor.invoke({"input": "Tell me a random cat fact!"})
print(result["output"])
````

````python
result = executor.invoke({"input": "List 5 cat breeds for me."})
print(result["output"])
````

````python
result = executor.invoke(
    {"input": "Give me 3 interesting cat facts, but keep them short (under 100 characters each)."}
)
print(result["output"])
````
