# Screenshotlayer Capture Demo

This notebook demonstrates how to use the Screenshotlayer MCP tool with a LangChain agent to capture website screenshots.

```python
import os

os.environ["SCREENSHOTLAYER_ACCESS_KEY"] = "your_api_key_here"

from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_openai import ChatOpenAI

from pymcpx.screenshotlayer import MCP_TOOLS
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
agent = create_tool_calling_agent(llm, MCP_TOOLS)
executor = AgentExecutor(
    agent=agent,
    tools=MCP_TOOLS,
    verbose=True,
    handle_parsing_errors=True,
)
```

```python
response = executor.invoke({"input": "Capture a screenshot of https://www.python.org with full page, 1920x1080 viewport, as PNG"})
print(response["output"])
```
