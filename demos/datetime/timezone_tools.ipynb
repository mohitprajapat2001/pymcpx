```python
from dotenv import load_dotenv
load_dotenv()

from demos.shared.llm_setup import get_llm
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from pymcpx.datetime import ConvertTimezoneTool, GetCurrentTimeTool

llm = get_llm()

tools = [GetCurrentTimeTool(), ConvertTimezoneTool()]

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a time and timezone assistant. Use the datetime tools to "
        "answer questions about dates, times, and timezone conversions.\n\n"
        "Available tools:\n"
        "- get_current_time(timezone, format) \u2014 current time in a timezone\n"
        "- convert_timezone(datetime_str, from_tz, to_tz) \u2014 convert between timezones",
    ),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke({"input": "What is the current time in UTC?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "What time is it right now in New York (America/New_York) and Tokyo (Asia/Tokyo)?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "Convert 2026-06-24T12:00:00 from America/New_York to UTC."})
print(result["output"])
```

```python
result = executor.invoke({"input": "If it's 9 AM in London (Europe/London), what time is it in Los Angeles (America/Los_Angeles)?"})
print(result["output"])
```
