```python
from dotenv import load_dotenv
load_dotenv()

from demos.shared.llm_setup import get_llm
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from pymcpx.datetime import (
    DatetimeToTimestampTool,
    GetDayOfWeekTool,
    TimestampToDatetimeTool,
)

llm = get_llm()

tools = [TimestampToDatetimeTool(), DatetimeToTimestampTool(), GetDayOfWeekTool()]

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a time and date assistant. Use the datetime tools to convert "
        "between timestamps and date strings, and find days of the week.\n\n"
        "Available tools:\n"
        "- timestamp_to_datetime(timestamp, timezone, format) \u2014 Unix timestamp \u2192 datetime str\n"
        "- datetime_to_timestamp(datetime_str, timezone) \u2014 datetime str \u2192 Unix timestamp\n"
        "- get_day_of_week(date_str) \u2014 day name for a date",
    ),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke({"input": "What date and time does Unix timestamp 1782316800 represent in UTC?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "Convert '2026-12-25T00:00:00' in UTC to a Unix timestamp."})
print(result["output"])
```

```python
result = executor.invoke({"input": "What day of the week is December 25, 2026?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "If a Unix timestamp is 0, what datetime does that correspond to in UTC? Also, what day of the week is 1970-01-01?"})
print(result["output"])
```
