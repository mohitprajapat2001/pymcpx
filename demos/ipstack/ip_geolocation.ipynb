# IPstack — IP Geolocation

Demonstrates **lookup_ip**, **lookup_requester_ip**, and **bulk_lookup_ip** tools
using the [IPstack API](https://ipstack.com).

Requires an `IPSTACK_ACCESS_KEY` environment variable set in `.env`.

```python
from dotenv import load_dotenv
load_dotenv()

from demos.shared.llm_setup import get_llm
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from pymcpx.ipstack import BulkLookupIPTool, LookupIPTool, LookupRequesterIPTool

llm = get_llm()

tools = [LookupIPTool(), LookupRequesterIPTool(), BulkLookupIPTool()]

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are an IP geolocation assistant. Use the IPstack tools to "
        "answer questions about IP addresses and their locations.\n\n"
        "Available tools:\n"
        "- lookup_ip(ip_address, security=1) \u2014 geolocation for a single IP\n"
        "- lookup_requester_ip() \u2014 geolocation for the requester's own IP\n"
        "- bulk_lookup_ip(ip_addresses) \u2014 geolocation for up to 50 IPs at once",
    ),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke({"input": "Look up geolocation data for the IP address 134.201.250.155. Include security insights."})
print(result["output"])
```

```python
result = executor.invoke({"input": "What is my own IP address and where is it located?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "Look up geolocation data for these IP addresses: 8.8.8.8 and 1.1.1.1"})
print(result["output"])
```
