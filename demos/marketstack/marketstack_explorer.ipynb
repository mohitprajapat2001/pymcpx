# Marketstack — Financial Data Explorer

Demonstrates **EOD data**, **intraday prices**, **real-time quotes**, **ticker info**,
**exchanges**, **dividends**, **splits**, **commodities**, **bonds**, **ETFs**,
**company ratings**, **indices**, **currencies**, **timezones**, and **SEC EDGAR** tools
using the [Marketstack API](https://marketstack.com).

Requires a `MARKETSTACK_ACCESS_KEY` environment variable set in `.env`.

```python
from dotenv import load_dotenv

load_dotenv()

from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

from demos.shared.llm_setup import get_llm
from pymcpx.marketstack import MarketstackToolkit

llm = get_llm()

tools = MarketstackToolkit().get_tools()

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a financial data assistant. Use the marketstack tools to answer "
            "questions about stocks, markets, and companies.\n\n"
            "Available tools:\n"
            "- marketstack_eod(symbols, date, latest, exchange, date_from, date_to, sort, limit, offset) \u2014 End-of-Day stock data\n"
            "- marketstack_intraday(symbols, interval, date, latest, exchange, date_from, date_to, sort, limit, offset, after_hours) \u2014 intraday stock prices\n"
            "- marketstack_realtime_price(ticker, exchange) \u2014 real-time stock price\n"
            "- marketstack_commodities(commodity_name, date_from, date_to, frequency) \u2014 commodity prices\n"
            "- marketstack_company_ratings(ticker, date_from, date_to, rated) \u2014 analyst ratings\n"
            "- marketstack_splits(symbols, symbol, date_from, date_to, sort, limit, offset) \u2014 stock split history\n"
            "- marketstack_dividends(symbols, symbol, date_from, date_to, sort, limit, offset) \u2014 dividend history\n"
            "- marketstack_ticker_info(symbol, exchange, use_tickerinfo_endpoint, limit, offset) \u2014 ticker details\n"
            "- marketstack_list_tickers(search, exchange, limit, offset) \u2014 search/list tickers\n"
            "- marketstack_indices(index, limit, offset) \u2014 index listings\n"
            "- marketstack_exchanges(mic, endpoint_type, symbols, date, interval, date_from, date_to, sort, limit, offset, search) \u2014 exchange queries\n"
            "- marketstack_currencies(limit, offset) \u2014 supported currencies\n"
            "- marketstack_timezones(limit, offset) \u2014 supported timezones\n"
            "- marketstack_bonds(country, limit, offset) \u2014 government bond yields\n"
            "- marketstack_etfs(ticker, list_type, date_from, date_to, limit, offset) \u2014 ETF holdings\n"
            "- marketstack_edgar_cik(company_name, cik_code, limit, offset) \u2014 CIK lookup\n"
            "- marketstack_edgar_submissions(cik_code, cik_code_name, report_from, report_to, filing_from, filing_to, accession_number) \u2014 SEC submissions\n"
            "- marketstack_edgar_facts(cik_code, query_type, frame, unit) \u2014 company facts",
        ),
        ("human", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke(
    {"input": "Find tickers related to 'Apple' and show me details about AAPL."}
)
print(result["output"])
```

```python
result = executor.invoke(
    {
        "input": "Get the latest EOD data for AAPL, MSFT, and GOOGL. Include the most recent trading day."
    }
)
print(result["output"])
```

```python
result = executor.invoke(
    {"input": "List the supported stock exchanges and tell me how many there are."}
)
print(result["output"])
```

```python
result = executor.invoke(
    {
        "input": "Show me analyst ratings for AAPL and check if there are any stock splits or dividends for AAPL."
    }
)
print(result["output"])
```

```python
result = executor.invoke(
    {
        "input": "Find the CIK code for Microsoft Corporation using the EDGAR tool, then check what facts are available for that CIK."
    }
)
print(result["output"])
```

```python
result = executor.invoke(
    {
        "input": "What are the current government bond yields for the United States? Also list the supported currencies."
    }
)
print(result["output"])
```

```python
result = executor.invoke({"input": "Get the current commodity price for gold and crude oil."})
print(result["output"])
```

```python
result = executor.invoke(
    {"input": "List all available ETFs. Then list the major stock market indices."}
)
print(result["output"])
```
