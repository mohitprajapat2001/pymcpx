```python
from dotenv import load_dotenv
load_dotenv()

from demos.shared.llm_setup import get_llm
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from pymcpx.converter import ConvertLengthTool, ConvertWeightTool

llm = get_llm()

tools = [ConvertLengthTool(), ConvertWeightTool()]

prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are a unit conversion assistant. Use the converter tools to "
        "convert between different units of measurement.\n\n"
        "Available tools:\n"
        "- convert_length(value, from_unit, to_unit) \u2014 length conversion\n"
        "  Supported units: m, cm, mm, km, inch, ft, yard, mile\n"
        "- convert_weight(value, from_unit, to_unit) \u2014 weight conversion\n"
        "  Supported units: kg, g, mg, lb, oz",
    ),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke({"input": "Convert 5 meters to centimeters."})
print(result["output"])
```

```python
result = executor.invoke({"input": "How many inches are in 2 feet? And how many centimeters is that?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "Convert 10 kilometers to miles."})
print(result["output"])
```

```python
result = executor.invoke({"input": "Convert 2.5 kilograms to grams. Also convert 16 ounces to pounds."})
print(result["output"])
```
