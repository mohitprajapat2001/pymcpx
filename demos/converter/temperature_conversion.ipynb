```python
from dotenv import load_dotenv

load_dotenv()

from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

from demos.shared.llm_setup import get_llm
from pymcpx.converter import ConvertTemperatureTool

llm = get_llm()

tools = [ConvertTemperatureTool()]

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a temperature conversion assistant. Use the tool to convert "
            "between Celsius, Fahrenheit, and Kelvin.\n\n"
            "Available tool:\n"
            "- convert_temperature(value, from_unit, to_unit) \u2014 temperature conversion\n"
            "  Units: c (Celsius), f (Fahrenheit), k (Kelvin)\n"
            "  Also accepts full names: celsius, fahrenheit, kelvin",
        ),
        ("human", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
print("\u2705 Agent ready. Running scenarios...")
```

```python
result = executor.invoke({"input": "Convert 0 degrees Celsius to Fahrenheit and to Kelvin."})
print(result["output"])
```

```python
result = executor.invoke({"input": "What is 98.6 degrees Fahrenheit in Celsius?"})
print(result["output"])
```

```python
result = executor.invoke({"input": "Convert 300 Kelvin to Celsius and to Fahrenheit."})
print(result["output"])
```

```python
result = executor.invoke({"input": "Convert -40 Celsius to Fahrenheit."})
print(result["output"])
```
