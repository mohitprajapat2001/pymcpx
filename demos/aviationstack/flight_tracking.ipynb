# Aviationstack Flight Tracking Demo

This notebook demonstrates how to use the Aviationstack MCP tools with a LangChain agent to track flights and retrieve aviation data.

```python
import os

os.environ["AVIATIONSTACK_ACCESS_KEY"] = "your_api_key_here"

from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_openai import ChatOpenAI

from pymcpx.aviationstack import MCP_TOOLS
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
agent = create_tool_calling_agent(llm, MCP_TOOLS)
executor = AgentExecutor(
    agent=agent,
    tools=MCP_TOOLS,
    verbose=True,
    handle_parsing_errors=True,
)
```

```python
response = executor.invoke({"input": "Show me all flights from JFK to LAX on 2025-09-15"})
print(response["output"])
```

```python
response = executor.invoke({"input": "Find airport information for JFK and LHR"})
print(response["output"])
```

```python
response = executor.invoke({"input": "Show me departure schedules for JFK airport"})
print(response["output"])
```
