# Zenserp SERP Search Demo

This notebook demonstrates how to use the Zenserp MCP tools with a LangChain agent to perform SERP searches across Google, YouTube, Bing, and more.

```python
import os

os.environ["ZENSERP_API_KEY"] = "your_api_key_here"

from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_openai import ChatOpenAI

from pymcpx.zenserp import MCP_TOOLS
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
agent = create_tool_calling_agent(llm, MCP_TOOLS)
executor = AgentExecutor(
    agent=agent,
    tools=MCP_TOOLS,
    verbose=True,
    handle_parsing_errors=True,
)
```

```python
response = executor.invoke({"input": "Search Google for the latest AI news from the past week"})
print(response["output"])
```

```python
response = executor.invoke({"input": "Find trending searches on Google right now"})
print(response["output"])
```

```python
response = executor.invoke({"input": "Search YouTube for machine learning tutorials"})
print(response["output"])
```
